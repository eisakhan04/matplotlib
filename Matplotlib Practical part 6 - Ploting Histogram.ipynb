# Ploting Histograms

```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
py_student_age = np.random.randint(23,45,(100))
ml_student_age = np.random.randint(15,40,(100))
```

```python
print(py_student_age)
print(ml_student_age)
```

```python
plt.hist(py_student_age)
plt.title("Python student age Histogram")
plt.xlabel('student age cotegory')
plt.ylabel('number student age')
plt.show()
```

```python
bins=[20,25,30,35,40,45,50]
plt.hist(ml_student_age)
plt.title('ML Student age ')
plt.xlabel("student age cotagory")
plt.ylabel("no student age")
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]

plt.hist(ml_student_age, bins, rwidth=0.1)

plt.title("ML Students age histograms")
plt.xlabel("Students age cotegory")
plt.ylabel("No. Students age")

plt.show()
```

```python
bins = [15,20,25,30,35,40,45]

plt.hist(ml_student_age, bins, rwidth=0.7, histtype = "bar")

plt.title("ML Students age histograms")
plt.xlabel("Students age cotegory")
plt.ylabel("No. Students age")

plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.figure(figsize= (18 ,11))
#plt.figure(figsize = (16,9))
plt.hist(ml_student_age, bins, rwidth=0.7, histtype = "bar"  , orientation='horizontal', color = "c", label = "ML Student")

plt.title("ML Students age histograms")
plt.xlabel("Students age cotegory")
plt.ylabel("No. Students age")
plt.legend(loc=1)

plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.figure(figsize = (16,9))
plt.hist(ml_student_age, bins, rwidth=0.8, histtype = "bar", orientation='vertical', color = "m", label = "ML Student")

plt.hist(py_student_age, bins, rwidth=0.8, histtype = "bar",orientation='vertical', color = "y", label = "Py Student")

plt.title("ML & Py Students age histograms")
plt.xlabel("Students age cotegory")
plt.ylabel("No. Students age")
plt.legend()
plt.show()
```

```python
from matplotlib import style
style.use("ggplot")
plt.figure(figsize = (16,9))

plt.hist([ml_student_age, py_student_age], bins, rwidth=0.8, histtype = "bar",
         orientation='vertical', color = ["m", "y"], label = ["ML Student", "Py Student"])

#plt.hist(py_students_age, bins, rwidth=0.8, histtype = "bar",
#         orientation='vertical', color = "y", label = "Py Student")

plt.title("ML & Py Students age histograms")
plt.xlabel("Students age cotegory")
plt.ylabel("No. Students age")
plt.legend()
plt.show()
```
