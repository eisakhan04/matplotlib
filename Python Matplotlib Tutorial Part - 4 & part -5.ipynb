**Example 11**

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(16, 9))

classes_index = [1, 2, 3]  # Example class indices
class_1 = [20, 25, 30]      # Example student counts for class 1
class_2 = [15, 18, 22]      # Example student counts for class 2
class_3 = [10, 12, 15]      # Example student counts for class 3

width = 0.2

plt.bar(classes_index, class_1, width, color="b", label="Class 1 Students")
plt.bar([index + width for index in classes_index], class_2, width, color="g", label="Class 2 Students")
plt.bar([index + width*2 for index in classes_index], class_3, width, color="y", label="Class 3 Students")

plt.title("Bar Chart of IAIP Class", fontsize=18)
plt.xlabel("Classes", fontsize=15)
plt.ylabel("No. of Students", fontsize=15)
plt.legend()
plt.show()
```

**Example 12**

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(16, 9))

classes_index = np.arange(1, 6)  # Assuming 5 classes
class_1 = [20, 25, 30, 18, 22]   # Example student counts for class 1
class_2 = [15, 18, 22, 12, 16]   # Example student counts for class 2
class_3 = [10, 12, 15, 8, 9]     # Example student counts for class 3

width = 0.2

plt.bar(classes_index, class_1, width, color="b", label="Class 1 Students")
plt.bar(classes_index + width, class_2, width, color="g", label="Class 2 Students")
plt.bar(classes_index + width + width, class_3, width, color="y", label="Class 3 Students")

plt.title("Bar Chart of IAIP Class", fontsize=18)
plt.xlabel("Classes", fontsize=15)
plt.ylabel("No. of Students", fontsize=15)
plt.legend()
plt.show()
```
