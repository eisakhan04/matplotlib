```python
classes =[ "Python", 'R', 'Machine Learning', 'Artificial Intelligence','Data Sciece']
class_1_student =[50,60,70,80,90]
#explode=[1,1,1,1,1]
explode = [0.03,0,0.1,0.0,0]
#colors = ["c", 'b','r','y','g']
textprops = {"fontsize":20}

plt.pie(class_1_student, labels= classes, explode=explode,colors=['c','b','r','y','g'] ,autopct = "%0.3f%%" ,
        shadow = True ,radius = 1.4, startangle = 270 ,textprops =textprops)
plt.show()
```

```python
plt.pie(class_1_student,labels = classes,explode = explode, colors = ['c','b','r','y','g'], autopct = "%0.2f%%",
        pctdistance = 0.7,shadow =True)
plt.show()
```

```python
plt.pie(class_1_student,
        labels = classes,
        explode = explode,
        colors = ['c','b','r','y','g'],
        autopct = "%0.2f%%",
        pctdistance = 0.5,
        shadow =True,
        labeldistance = 1.6,
        startangle = 270)
plt.show()
```

```python
colors = ['c','b','r','y','g']
plt.pie(
        class_1_student,
        labels = classes,
        explode = explode,
        colors = colors,
        autopct = "%0.2f%%",
        pctdistance = 0.6,
        shadow =True,
        labeldistance = 1.6,
        startangle = 270,
        radius = 1,
        counterclock = True)
plt.show()
```

```python
explode = [0.03,0,0.1,0,0]
colors = ["c", 'b','r','y','g']
textprops = {"fontsize":15}

plt.pie(class_1_student, labels = classes, explode = explode,
        colors =colors, autopct = "%0.2f%%", shadow = True, radius = 1.4,
       startangle = 270, textprops =textprops)
plt.legend(loc='center')
plt.show()
```
